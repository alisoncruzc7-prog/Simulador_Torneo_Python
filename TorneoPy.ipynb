**Que comienze el Torneo eSport¬°¬° ‚öΩ**

```python
jugadores = ["Player1", "Player2", "Player3"]
puntajes = [80, 40, 85]
kills = [5, 2, 8]
asistencias = [3, 1, 4]
```

```python
puntajes = [calcular_puntaje(k, a) for k, a in zip(kills, asistencias)]
```

```python
def calcular_puntaje(k, a):
    return k * 10 + a * 5
```

```python
clasificacion = ["Pasa" if p >= 50 else "No pasa" for p in puntajes]
```

```python
for i in range(len(jugadores)):
    kills[i] += 1
    asistencias[i] += 1
    puntajes[i] = calcular_puntaje(kills[i], asistencias[i])
```

```python
import pandas as pd
data = {
    "Jugador": jugadores,
    "Kills": kills,
    "Asistencias": asistencias,
    "Puntaje": puntajes,
    "Clasificaci√≥n": clasificacion
}
df = pd.DataFrame(data)
```

**Resultados Finales...**
